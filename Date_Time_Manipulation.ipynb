```python
df2['Date']=df2['DateTime'].str.split(' ').str[0]
```

```python
df2.head()
```

```python
#df2['Time'] = pd.to_datetime(df2['Time']).strftime('%H:%M:%S')
df2.info()
```

```python
df2['Time_numeric']=df2.Time.str.extract('(\d+)')
```

```python
df2['Time_AM_PM']=df2.Time.str[-2:]
```

```python
df2['Time_numeric']=df2['Time_numeric'].astype('int')
```

```python
df2['Time_numeric']=np.where(df2['Time_AM_PM']=='AM',df2['Time_numeric'],df2['Time_numeric']+12)
```

```python
df2['Time_numeric']=df2['Time_numeric'].astype('str')
```

```python
df2['Date']=df2['Date'] + " " +df2['Time_numeric']
```
